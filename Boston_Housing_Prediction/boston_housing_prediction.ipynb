# Boston Housing Price Prediction

```python
# Importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

import pickle
```

```python
# Loading the data
boston_df = pd.read_csv("BostonHousing.csv", encoding='latin-1')

# Viweing the columns of the data
boston_df.head()
```

```python
# Renaming the target
boston_df.rename(columns = {'medv':'price'}, inplace = True)

# Viweing the columns of the data
boston_df.head()
```

### Exploratory Data Analysis

```python
# Checking for missing values in the data
boston_df.isnull().sum()
```

```python
# Checking how different features are correlated with each other
boston_df.corr()
```

```python
# Plotting a regression line for a feature that is highly correlated with the target
sns.regplot(data=boston_df, x="rm", y="price")
```

### Model Training

```python
# Seggregating the data and the target
X = boston_df.iloc[:,:-1]
y = boston_df.iloc[:,-1]

# Creating an object for standardization of data
scaler = StandardScaler()
```

```python
# Creating a function to perform Linear Regression
def Linear_Regression(X, y, random_state):

    # Splitting the data into train and test data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=random_state)

    # Standardizing the train and test data to make sure features are in the same scale for gradient descent
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    # Creating a linear regression model
    linear_model = LinearRegression()

    # Fitting the model to the train data
    linear_model.fit(X_train, y_train)

    # Predicting the target values for the test data
    y_pred = linear_model.predict(X_test)

    # Calculating the R2 score
    score = r2_score(y_test, y_pred)
    
    # Returning the R2 score
    return score, linear_model

# Creating a function to find the most efficient model
def Efficient_Linear_Regression(X, y):
    
    # Initializing the max score and best random state to 0
    max_score = 0
    best_state = 0
    best_model = None
    
    # Creating a loop from 0 to 99
    for i in range(0, 100):
        
        # Performing Linear Regression with each random state
        score, linear_model = Linear_Regression(X, y, random_state=i)
        
        # Checking whether current score is better than max score
        if(score > max_score):
            
            # Assigning current score to max score and current model to best model
            max_score = score
            best_model = linear_model
            
    # Returning the max score and the best linear model
    return max_score, best_model
```

```python
# Performing efficient Linear Regression
score, linear_model = Efficient_Linear_Regression(X, y)

# Displaying the score
score
```

### Predicting new data

```python
# Conisdering a new data point
new_data = boston_df.iloc[:,:-1].values[0].reshape(1,-1)

# Standardizing the new data point
new_data = scaler.transform(new_data)

# Predicting the price of new data point
linear_model.predict(new_data)
```

### Saving the model

```python
# Saving the standardizer
pickle.dump(scaler, open("scaler.pkl", "wb"))

# Saving the model
pickle.dump(linear_model, open("linear_model.pkl", "wb"))
```
